```python
# Import math Library
import math
# Import random library
import random
# Import matplotlib.pyplot to plot
import matplotlib.pyplot as plt

# The value of pi
PI = math.pi
# The value of Radius
RADIUS = 1

# Generate random points of (x,y)
def generatePoint():
    x = random.random() #[0,1)
    y = random.random() #[0,1)
    return x, y

def errorRate(estPi):
    # PI is the actual Pi and estPi is the estimation
    return (abs((PI - estPi)/PI) * 100)
    
def plotResult(inCircle, outCircle):
    zip_in = zip(*inCircle)
    zip_out = zip(*outCircle)
    plt.scatter(*zip_in, marker='.', color='b')
    plt.scatter(*zip_out, marker='.', color='r')
    plt.xlabel('Figure 1: Estimating  using simulation (10^4)')
    plt.show()
    
def main():
    # Save the results for output
    numOfPointStr = ['10 ^ 3', '10 ^ 4', '10 ^ 5', '10 ^ 6']
    resultPi = []
    resultErr = []
    
    # Save the result for plotting
    plotIn = []
    plotOut = []
    
    # Number of points for each run
    numOfPoints = [10**3, 10**4, 10**5, 10**6]
    
    # For each n in {10^3, 10^4, 10^5, 10^6}
    for points in numOfPoints:
        #reset the set of point
        inCircle = []
        outCircle = []
        
        for point in range(points):
            # Generate the point
            x, y = generatePoint()

            # Calculate the distance
            distance = math.sqrt(x**2 + y**2)

            # Determine where the point is
            if(distance < RADIUS):
                inCircle.append((x, y))
            else:
                outCircle.append((x, y))
                
        # If n = 10^4, save the set of point
        if(points == 10**4):
            plotIn = inCircle
            plotOut = outCircle
                
        # Esitmate Pi
        tempPi = 4 * len(inCircle) / points
        # Add pi to the result list
        resultPi.append(tempPi)
        
        # Calculate the error
        tempErr = errorRate(tempPi)
        # Add error to the list
        resultErr.append(tempErr)
        
    # Print out the result
    for i in range(len(numOfPoints)):
        print(f'n = {numOfPointStr[i]}', 'pi = {:.6f} error = {:.6f} %'.format(resultPi[i], resultErr[i]))
    plotResult(plotIn, plotOut)
    
main()
```
